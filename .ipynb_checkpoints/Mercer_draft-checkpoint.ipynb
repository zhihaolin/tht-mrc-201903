# Question 1

Can we conduct a descriptive analysis of attrition trends over the past three years for Production Assemblers? I would like to understand if there are any patterns in the attrition data and whether or not any particular employee groups are more prone to attrition.

```python
#Find all Production Assemblers in Data

data_pa_orig = data[(data['JobTitle']=="Production Assembler I")|(data['JobTitle']=="Production Assembler II")]

#Keep only the latest entry for every employee
data_pa=data_pa_orig.loc[data_pa_orig.groupby("EmployeeID").EffectiveDate_x.idxmax(),:]
print("Number of PAs in data:" + str(data_pa.shape[0]))

#Create a new column of termination date
data_pa['TerminationDate_C'] = data_pa["TerminationDate_y"]

#There are 7 PA with status "Terminated" but no termination date in Demographics spreadsheet
data_pa.loc[(data_pa["Status"]=="Terminated") & (np.isnat(data_pa["TerminationDate_y"])),"TerminationDate_C"] = data_pa["TerminationDate_x"]

#Extract termination year and termination month
data_pa['Termination_year'] = data_pa["TerminationDate_C"].dt.year
data_pa['Termination_month'] = data_pa["TerminationDate_C"].dt.month

#There are 7 PA with status "Terminated" but no termination date in Demographics spreadsheet
#data_pa.loc[(data_pa["Status"]=="Terminated") & (np.isnan(data_pa["Termination_year"])),"Termination_year"] = data_pa["TerminationDate_x"].dt.year
#data_pa.loc[(data_pa["Status"]=="Terminated") & (np.isnan(data_pa["Termination_month"])),"Termination_month"] = data_pa["TerminationDate_x"].dt.month

#Correct the status of employment
data_pa["Status_C"] = data_pa["Status"]

#There are 13 PA with status "Active" but is already terminated (based on Demographics)
data_pa.loc[(data_pa["Status"]=="Active") & np.invert(np.isnan(data_pa["Termination_year"])),"Status_C"] = "Terminated"

#Find length of employment
data_pa['Length_employment'] = data_pa["TerminationDate_C"] - data_pa["StartDate"]
data_pa.loc[np.invert(np.isnat(data_pa["RehireDate"])), "Length_employment"] = data_pa["TerminationDate_C"] - data_pa["RehireDate"]
data_pa['Length_employment'] = data_pa['Length_employment'].dt.days
print("Number of terminated PAs in data:" + str(np.sum(data_pa["Status_C"]=="Terminated")))

#Extract start year and start month
data_pa['Start_year'] = data_pa["StartDate"].dt.year
data_pa['Start_month'] = data_pa["StartDate"].dt.month

#Fill in
data_pa["EmployeeClass"].fillna("A", inplace=True)
```

# Question 2

Can we predict the likelihood that a Production Assembler will leave Mercer Manufacturing in advance of the event, to allow the company sufficient time to intervene? I would like to prompt my Managers and alert them of high risk individuals

```python
data_pa_pred = data_pa.loc[:, ["DepartmentID", "ManagerID", "EmployeeClass", "Compensation", "Age",
                        "Gender", "Ethnicity", "Country", "Status_C"]]
```

```python
le = preprocessing.LabelEncoder()

for col in ["DepartmentID", "ManagerID", "EmployeeClass", "Gender", "Ethnicity", "Country", "Status_C"]:
    n = len(data_pa_pred[col].unique())
    print(col)
    
    if (n > 2):
       X = pd.get_dummies(data_pa_pred[col])
       X = X.drop(X.columns[0], axis=1)
       data_pa_pred[X.columns] = X
       data_pa_pred.drop(col, axis=1, inplace=True)  # drop the original categorical variable (optional)
    else:
       le.fit(data_pa_pred[col])
       data_pa_pred[col] = le.transform(data_pa_pred[col])
```

```python
data_pa_pred.shape
```

## Logistic regression

```python
X_train, X_test, y_train, y_test = train_test_split(
...     data_pa_pred.loc[:, data_pa_pred.columns != 'Status_C'], data_pa_pred.loc[:, 'Status_C'], test_size=0.33, random_state=0)
```

```python
print(sum(y_train==1))
print(sum(y_train==0))
print(sum(y_test==1))
print(sum(y_test==0))
```

```python
#Status_C

#Department ID
#Manager ID
#EmployeeClass
#Compensation
#Age
#Gender
#Ethnicity
#Country

#State
#Performance
#Distance to factory
#Postal code
#Start or rehire date year and month
```

```python
LogReg = LogisticRegression()
logreg = LogReg.fit(X_train, y_train)
#y_pred = LogReg.predict(X_test)
y_pred = logreg.predict(X_test)
#print(y_test.shape)
#print(y_pred.shape)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
print("Training set score: {:.3f}".format(logreg.score(X_train,y_train)))
print("Test set score: {:.3f}".format(logreg.score(X_test,y_test)))
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
#confusion_matrix
print(precision_score(y_test, y_pred))
print(recall_score(y_test, y_pred))
```
